## DDMS data exploration

```python
import os
import dicom
import numpy as np
import cv2

failcount=0
sucesscount=0
def dicom2img(dicom_file, img_file):

    # Extracting data from the mri file
    plan = dicom.read_file(dicom_file)
    #print(plan.pixel_array.shape)
    amax = np.amax(plan.pixel_array)
    image_2d_scaled = plan.pixel_array / amax * 255.0
    cv2.imwrite(img_file, image_2d_scaled)

# traverse root directory, and list directories as dirs and files as files
def walk_dir(input_dir,output_dir):  
    for root, dirs, files in os.walk(input_dir,output_dir):
        path = root.split(os.sep)
        for file in files:
            src_path=root+os.sep+file
            if os.path.isfile(src_path):
                dst_path=root.replace(input_dir+os.sep,output_dir+os.sep)
                if not os.path.exists(dst_path):
                    os.makedirs(dst_path)
                dst_file=src_path.replace(input_dir+os.sep,output_dir+os.sep)
                filename=file.split('.')[0]+'.jpg'
                #print(filename)
                try:
                    dicom2img(src_path,dst_path+os.sep+filename)
                    global sucesscount
                    sucesscount+=1
                except Exception:
                    global failcount
                    failcount+=1
                    print('fail to convert this dicom',src_path)
```

# Write file to target directory

```python
# traverse root directory, and list directories as dirs and files as files
import re
failcount=0
sucesscount=0

def preprocess_dicom(input_dir,output_dir):  
    for root, dirs, files in os.walk(input_dir):
        path = root.split(os.sep)
        for file in files:
            src_path=root+os.sep+file
            if os.path.isfile(src_path):

                try:
                    result = re.search('P_\d{5}_(LEFT|RIGHT)_CC', src_path)
                    if result:
                        filename=result.group(0)

                        print('===save image as name(',filename,')')
                        clazz=id_class.get(filename)
                        tgt_path=output_dir+os.sep+clazz
                        if not os.path.exists(tgt_path):
                            os.makedirs(tgt_path)
                        full_tgt_file_name=tgt_path+os.sep+filename+'.jpg'
                        print(full_tgt_file_name)
                        dicom2img(src_path,full_tgt_file_name)
                    global sucesscount
                    sucesscount+=1
                except Exception:
                    global failcount
                    failcount+=1
                    #print('fail to convert this dicom',src_path)
```
